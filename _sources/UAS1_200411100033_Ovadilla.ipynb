```python
import pandas as pd
import requests
import csv
from bs4 import BeautifulSoup
```

# **Crawling Berita**

Disini saya melakukan crawling berita dalam 3 kategori yang berbeda. Untuk proses crawling yang saya lakukan ini, link berita yang dimasukkan juga sesuai dengan link yang ada pada kategori. Kategori yang saya ambil yaitu :
1. Politik
2. Nasional
3. Food (Resep Masakan)

Cara crawling yang saya lakukan disini yaitu dengan mengambil judul dan isi beritanya saja. Untuk mengambil judul dan isi berita disini dapat diproses melalui class yang ada pada elemen web berita online.

Setiap kategori memiliki halaman berita yang berbeda-beda. Jadi, dapat dimasukkan pada variabel n yang sudah saya sediakan.

## Crawling Berita POLITIK

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

linkberita = 'https://sorotpolitik.kompas.com/'
listpolitik = []
n = 41

for halaman in range(1, n+1):
  req = requests.get(linkberita)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('div', 'article__list clearfix') #class semua berita
  for it in items:
    link = it.find('a','article__link')['href'] #class perberita

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judul = soup2.find('h1', 'read__title').text    #class judul berita
    isi = "".join(soup2.find('div', 'read__content').text.split('\n'))

    listpolitik.append([judul,isi])
```

## Crawling Berita NASIONAL

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

linkberita = 'https://nasional.kompas.com/search/'
listt = []
n = 300

for halaman in range(1, n+1):
  req = requests.get(linkberita)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('div', 'article__list clearfix') #class semua berita
  for it in items:
    link = it.find('a','article__link')['href'] #class perberita

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judul = soup2.find('h1', 'read__title').text    #class judul berita
    isi = "".join(soup2.find('div', 'read__content').text.split('\n'))

    listt.append([judul,isi])
```

Menggabungkan 2 data crawling berita politik dan nasional menjadi 1 data frame dan memberikan label.

```python
politik = pd.DataFrame(listpolitik, columns=['Judul', 'Isi'])
politik['Label'] = 'Politik'

nasional = pd.DataFrame(listt, columns=['Judul', 'Isi'])
nasional['Label'] = 'Nasional'

gabung = [politik, nasional]
combined_df = pd.concat(gabung, ignore_index=True)
combined_df
```

```python
combined_df.to_csv('PolitikNasional.csv',  index=False)
```

## Crawling Berita FOOD

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

urlberita = 'https://www.kompas.com/food/resep'
beritaFood = []
n = 10

for halaman in range(1, n+1):
  req1 = requests.get(urlberita)
  soup1 = BeautifulSoup(req1.text, 'html.parser')
  items = soup1.findAll('div', 'foodLatest__item clearfix') #class berita
  for it in items:
    link = it.find('a','food__link')['href'] #class perberita (pada bagian judul)

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judulF = soup2.find('h1', 'read__title').text    #class judul berita
    isiF = "".join(soup2.find('div', 'read__content').text.split('\n'))

    beritaFood.append([judulF,isiF])
```

## DataFrame 3 Kategori

Menggabungkan crawling berita politik, nasional, dan food menjadi 1 data frame dan sudah ada labelnya masing-masing.

```python
food = pd.DataFrame(beritaFood, columns=['Judul', 'Isi'])
food['Label'] = 'Food'

gabungsemua = [combined_df, food]
keseluruhan = pd.concat(gabungsemua, ignore_index=True)
keseluruhan
```

```python
keseluruhan.to_csv('Berita3Kategori.csv',  index=False)
```
